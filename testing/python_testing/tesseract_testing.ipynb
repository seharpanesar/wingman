```python
import pytesseract
# If you don't have tesseract executable in your PATH, include the following:
pytesseract.pytesseract.tesseract_cmd = r'/opt/homebrew/Cellar/tesseract/5.3.3/bin/tesseract'
# Importing Image and ImageOps module from PIL package
from PIL import Image, ImageOps
import os 
import time 

directory_path = os.getcwd() + "/images"

for filename in os.listdir(directory_path):
    # Check if the item is a file
    if os.path.isfile(os.path.join(directory_path, filename)):
        # creating a image1 object
        im1 = Image.open(f"./images/{filename}")
        
        # applying grayscale method
        im2 = ImageOps.grayscale(im1)

        starttime = time.time()
        # Simple image to string
        extracted = pytesseract.image_to_string(im2)
        endtime = time.time()
        execution_time = endtime - starttime

        print(extracted)
        print(f"Time for {filename}: {execution_time}")
    
     
     
```
