```python
import matplotlib.pyplot as plt

# data from tests
x = [1, 2, 3, 4, 5]  
server= [0.136, 0.241, 0.226, 0.298, 0.316 ]
local = [0.254, 0.328, 0.302, 0.435, 0.479]

# line plot
plt.plot(x, server, label='Server OCR time')
plt.plot(x, local, label='Local OCR time')

# labels
plt.xlabel('Figure number')
plt.ylabel('Time (sec)')
plt.title('Server OCR Time vs Local OCR Time')

# limiting x-axis from 1 to 5
plt.xlim(1, 5)
plt.xticks(range(1, 6))
plt.legend()
plt.show()
```
